```python
import pandas as pd

file_path = r"C:\Desarrollo\core\test\input\SI - ODS - Detalle de reclamo por siniestro.xlsx"

try:
    df_domains = pd.read_excel(file_path)
    print("üìÑ Data cargada correctamente:")
    print(df_domains.head())  # Muestra las primeras filas para verificar
except Exception as e:
    print(f"‚ùå Error al leer el archivo: {e}")
```

```python
df_domains.columns

df_filtered = df_domains[['Fecha y hora de ocurrencia del siniestro', 'Patente  del veh√≠culo damnificado','Patente del veh√≠culo del asegurado']]
```

```python
# Crear la columna 'Domain' basada en la comparaci√≥n de patentes
df_filtered['Domain'] = df_filtered.apply(lambda row: 
    row['Patente  del veh√≠culo damnificado'] if pd.notna(row['Patente  del veh√≠culo damnificado']) else row['Patente del veh√≠culo del asegurado'], 
    axis=1)

df_filtered
```

```python
# Filtrar solo registros donde ambas patentes sean iguales o donde una est√© en null
df_filtered = df_filtered[
    (df_filtered['Patente  del veh√≠culo damnificado'].fillna('') == df_filtered['Patente del veh√≠culo del asegurado'].fillna(''))
]
```

```python
df_filtered
```

```python
# Filtrar solo registros donde ambas patentes sean iguales o donde una est√© en null
df_filtered = df_filtered[
    (df_filtered['Patente  del veh√≠culo damnificado'].fillna('') == df_filtered['Patente del veh√≠culo del asegurado'].fillna(''))
]

# Seleccionar solo las columnas necesarias
df_result = df_filtered[['Fecha y hora de ocurrencia del siniestro', 'Domain']]

df_result
```

```python
# Eliminar filas con NaN en 'Domain'
df_result_clean = df_result.dropna()
df_result_clean
```

```python
duplicados = df_result_clean[df_result_clean.duplicated()]
print(duplicados)
```

```python
duplicados = duplicados.drop_duplicates()
duplicados
```

```python
# Guardar el DataFrame en un archivo CSV
duplicados.to_csv( r"C:\Desarrollo\pipeline-master\test\output\limpieza_siniestro_pipeline.csv", index=False, encoding='utf-8')

print("Archivo CSV generado: limpieza_siniestro_pipeline.csv")
```
