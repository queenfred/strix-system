```python
import pandas as pd
from security.services.health_check import health_check
from core.services.vehicle_service import VehicleService
from core.services.portfolio_service import PortfolioService
#from services.portfolio_event_service import PortfolioEventService
from core.services.domain_service import DomainService

from datetime import datetime, timedelta

file_path = "C:/Desarrollo/strix-system/pipeline/test/output/limpieza_siniestro_pipeline.csv"
vehicle_service = VehicleService()
portfolio_service = PortfolioService()
domain_service = DomainService()


print("üì• Cargando datos del archivo CSV...")
df = pd.read_csv(file_path, encoding="utf-8")
df["Domain"] = df["Domain"].astype(str).str.strip()
df["Fecha y hora de ocurrencia del siniestro"] = pd.to_datetime(df["Fecha y hora de ocurrencia del siniestro"], dayfirst=True)

domain_list = df["Domain"].tolist()

print(f"üîé Buscando {len(domain_list)} dominios en strix.vehicle...")
vehicle_service.create_domains_from_vehicles(domains=domain_list)
```

```python
# Obtener los registros reales desde la tabla domain
print("üîÑ Resolviendo nombres de dominio a IDs reales...")
existing_domains = vehicle_service.domain_repo.get_domains_by_names(domain_list)  # M√©todo que ten√©s que tener
domain_name_to_id = {d.domain: d.id for d in existing_domains}

```

```python
# Filtramos los dominios que realmente existen y extraemos los IDs
valid_domain_ids = list(domain_name_to_id.values())

print(f"‚úÖ Se encontraron {len(valid_domain_ids)} dominios v√°lidos con ID.")

```

```python
# Ahora s√≠, crear el portfolio con los IDs v√°lidos
print("üìÇ Creando un nuevo portfolio con estos dominios...")
portfolio_id = portfolio_service.create_portfolio_with_domains("Siniestro Portfolio", valid_domain_ids)
```

```python
print(f"‚úÖ Portfolio {portfolio_id} creado exitosamente.")
```
